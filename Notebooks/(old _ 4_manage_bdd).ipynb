Des dosssiers sont nommés avec le format suivant :
Lasioglossum (Evylaeus) punctatissimum

On enlève le sous genre entre parenthèses

```python
import re 

def rename_folders(path): 

    for folder in os.listdir(path): 

        # look if parenthesis
        if '(' in folder :

            # get new name i.e without parenthesis and content
            new_name = re.sub(r'\([^)]*\)', '', folder)

            # remove if two consecutive spaces
            new_name = re.sub(' +', ' ', new_name)

            # get old path
            old_path = os.path.join(path,folder)

            # get new path
            new_path = os.path.join(path,new_name)

            # rename folder
            os.rename(old_path,new_path)
```

```python
def move_to_clean(path): 


    # loop over databases 

    for db in os.listdir(path): 

        db_path = os.path.join(path,db)

        # loop over labels
        for label in os.listdir(db_path):

            label_path = os.path.join(db_path,label)

            # create dir if not exists
            if not os.path.exists(os.path.join(path_to_whole_dataset_cleaned,label)):
                os.mkdir(os.path.join(path_to_whole_dataset_cleaned,label))

            # loop over files
            for file in os.listdir(label_path): 

                file_path = os.path.join(label_path,file)

                new_name = db + '_' + file

                # copy file
                new_path = os.path.join(path_to_whole_dataset_cleaned,label,new_name)

                cmd = 'cp "' + file_path + '" "' + new_path + '"'
                os.system(cmd)
```

# 4. Editions de csv récapitulatifs 



```python
# edit csv file to sum up strucutre of dataset

import os 
import pandas as pd 

def edit_csv(path,output_path): 
    # csv must have columns : 
    # path,label

    # create dict
    dict = {}

    # loop over files
    for root,dirs,files in os.walk(path):

        for file in files: 

            # get label
            label = root.split('/')[-1]

            # get path
            file_path = os.path.join(root,file)

            # add to dict
            dict[file_path] = label

    # create df with path and label as columns
    df = pd.DataFrame.from_dict(dict,orient = 'index',columns = ['label'])

    # add path column
    df['path'] = df.index

    # drop index 
    df = df.reset_index(drop = True)

    # save to csv
    df.to_csv(output_path,errors = 'ignore')
```

```python
path_whole_dataset = "../data_bees_detection/whole_dataset"
path_whole_dataset_cleaned = "../data_bees_detection/whole_dataset_cleaned"


edit_csv(path_whole_dataset,"/datafiles/classification/whole_dataset_230628.csv")
edit_csv(path_whole_dataset_cleaned,"/datafiles/classification/whole_dataset_cleaned_230628.csv")
```

# 5. Filtrer le dataset pour créer un dataset de classification

Avec les contraintes suivantes : 

- Uniquement des images labelisées au niveau de l'espèce   
- Nombre d'images par espèces compris entre 50 et 10000 

(NB : cette dernière étape aurait pu se faire différement : 
pour les espèces avec plus de 1000 img; faire tourner yolo sur
toutes les images et prendre les 1000 avec le plus gros score 
de confiance )



```python
import pandas as pd 
import os 
```

```python
df_whole_dataset_cleaned = pd.read_csv("/datafiles/classification/datasets/whole_dataset_cleaned_230628.csv")

df_whole_dataset_cleaned.head()
```

```python
def accept_label(label): 

    # 1. are species

    # in one word
    if len(label.split(' ')) == 1 :
            
        return False

    # contains ' sp'
    elif ' sp' in label : 

        return False
    
    else : 

        return True
    
# remove labels that are not species
df_whole_dataset_cleaned['species'] = df_whole_dataset_cleaned['label'].apply(accept_label)

# remove labels that are not species
df_whole_dataset_cleaned = df_whole_dataset_cleaned[df_whole_dataset_cleaned['species'] == True]



# more than 50 images per label
df_whole_dataset_cleaned = df_whole_dataset_cleaned[df_whole_dataset_cleaned.groupby('label').label.transform('count') > 50]


```

```python
accept_label('Coelioxys')
```

```python
df_whole_dataset_cleaned['label'].value_counts().describe()
```

```python
labels = df_whole_dataset_cleaned['label'].unique()

# create dict
dict = {}

# loop over labels
for label in labels :

    # get label count
    label_count = df_whole_dataset_cleaned[df_whole_dataset_cleaned['label'] == label]['label'].count()

    # add to dict
    dict[label] = label_count

# create df
df = pd.DataFrame.from_dict(dict,orient = 'index',columns = ['label_count'])

# sort df
df = df.sort_values(by = 'label_count',ascending = False)

df.to_csv("/datafiles/classification/whole_dataset_cleaned_nb_img_per_labels_230628.csv")
```

```python
# cap number of pictures per label to 1000

df = df_whole_dataset_cleaned.copy()

# Group the dataframe by the '#label' column
grouped = df.groupby('label')

# Create an empty list to store the filtered data
filtered_data = []

# Iterate over each group
for label, group in grouped:
    if len(group) >= 1000:
        # If the group has 1000 or more paths, add only 1000 paths
        filtered_data.append(group.sample(1000))
        
    else:
        # If the group has fewer than 1000 paths, add all of them
        filtered_data.append(group)

# Concatenate the filtered data into a new dataframe
df_filtered = pd.concat(filtered_data)

# Reset the index of the filtered dataframe
df_filtered = df_filtered.reset_index(drop=True)

```

```python
import os

def new_path(path, output_path="../data_bees_detection/whole_dataset_cleaned_only_species_cap_1000"):
    
    # Get label
    label = path.split('/')[-2]

    new_path = os.path.join(output_path, label)
    
    return new_path



def move_to_folder(df, output_path, new_path_func):



    # Change path column
    df['new_path_desti'] = df['path'].apply(lambda x: new_path_func(x, output_path))

    # Loop over labels
    for label in df['label'].unique():

        print('Moving label: ', label)

        # Create directory if it does not exist
        if not os.path.exists(os.path.join(output_path, label)):
            os.mkdir(os.path.join(output_path, label))

        # Loop over files
        for i, file in enumerate(df[df['label'] == label]['path']):
            
            # Get new path
            new_path = df[df['label'] == label]['new_path_desti'].iloc[i]

            file_name = file.split('/')[-1]

            new_path = os.path.join(new_path, file_name)


            # Copy file
            cmd = 'cp "' + file + '" "' + new_path + '"'

            os.system(cmd)
```
